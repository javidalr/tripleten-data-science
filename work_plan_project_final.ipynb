## **1. Descripción del proyecto**

Al operador de telecomunicaciones Interconnect le gustaría poder pronosticar su tasa de cancelación de clientes. Si se descubre que un usuario o usuaria planea irse, se le ofrecerán códigos promocionales y opciones de planes especiales. El equipo de marketing de Interconnect ha recopilado algunos de los datos personales de sus clientes, incluyendo información sobre sus planes y contratos.

### **1.1 Servicios de Interconnect**

Interconnect proporciona principalmente dos tipos de servicios:

1. Comunicación por teléfono fijo. El teléfono se puede conectar a varias líneas de manera simultánea.
2. Internet. La red se puede configurar a través de una línea telefónica (DSL, *línea de abonado digital*) o a través de un cable de fibra óptica.

Algunos otros servicios que ofrece la empresa incluyen:

- Seguridad en Internet: software antivirus (*ProtecciónDeDispositivo*) y un bloqueador de sitios web maliciosos (*SeguridadEnLínea*).
- Una línea de soporte técnico (*SoporteTécnico*).
- Almacenamiento de archivos en la nube y backup de datos (*BackupOnline*).
- Streaming de TV (*StreamingTV*) y directorio de películas (*StreamingPelículas*)

La clientela puede elegir entre un pago mensual o firmar un contrato de 1 o 2 años. Puede utilizar varios métodos de pago y recibir una factura electrónica después de una transacción.

### **1.2 Descripción de los datos**

Los datos consisten en archivos obtenidos de diferentes fuentes:

- `contract.csv` — información del contrato;
- `personal.csv` — datos personales del cliente;
- `internet.csv` — información sobre los servicios de Internet;
- `phone.csv` — información sobre los servicios telefónicos.

En cada archivo, la columna `customerID` (ID de cliente) contiene un código único asignado a cada cliente. La información del contrato es válida a partir del 1 de febrero de 2020.

## **2. Preguntas e hipótesis**

1. Al examinar los conjuntos de datos dispuestos para el proyecto, el estudio de la tasa de abandono esta solo visto desde el ambito comercial y caractrísticas personales del cliente, falta tener retroalimentación de la parte técnica del servicio, en cuanto a calidad de servicio, tiempos de respuesta, atención al cliente, etc., ya que eso tambien afecta la tasa de cancelación.

2. El archivo `contract` tiene evidencia que los contratos comienzan desde el año 2013, pero los clientes que han cancelado su servicio, empiezan a partir desde octubre de 2019, por ello, que cambió en el segundo semestre del 2019 que como consecuencia tuvó el aumento de la tasa de abandono?.

3. Los clientes necesitan adquirir solo el servicio de internet, lo pueden hacer?, o se genera una "venta atada", de planes de telefono + internet, ya que pagar por un servicio que no se utilizará propicia el invremento de la tasa de perdida de clientes.

4. Los clientes firman un contrato por 1 o 2 años, tienen mejores condiciones comerciales que lo clientes que realizan el pago mensual?, afecta la tasa de abandono?

5.  Es un buen plateamiento aplicar cargos por soporte técnico?, siendo que la empresa es la que entrega el servicio, este item debiese estar incorporado en la tarifa basica de los servicios ofrecidos al cliente, a no ser que sea un soporte técnico especializado que ayude al cliente con problemas técnicos indirectos asociados a los servicios contratados. 

6. Las caracteristicas personales de cada cliente, influyen en la tasa de deserción de clientes?, ya que en los datos proporcionados existe certeza sobre su género, si es adulto mayor, si tiene pareja/esposa, hijos y/o alguien dependiente.

7. El tipo de facturación(Electrónica o física), los métodos de pago(Cheque por correo, cheque electrónico, transferencia o trajeta de crédito), el tipo de pago (Mensual, anual o bienal) intervienen en la tasa de abandono?

8. El tipo de servicio de internet(DSL o fibra óptica), los servicios adicionales(seguridad en línea, streaming de TV, streaming de péliculas, protección de dispositivos, etc) y las multiples líneas asociadas al cliente, afectan de alguna manera la tasa de cancelación?

9. Examinar la proporción de clientes activos vs clientes que han cancelado su suscripción, verificar si existe un desbalance de clases, en caso de existir, indagar y aplicar técnicas y/o métodos de sobremuestro o submuestreo al conjunto de datos.


## **3. Pasos a desarrollar**

1. Inicialmente se obtienen 4 archivos, los cuales es necesario revisar y cargar los datos en variables, asi dejarlos disponibles para el desarrollo del proyecto.

2. Carga y lectura de datos: Combinar variables con archivos y generar un solo conjunto de datos, mediante la columna `customer_id`, la cual todos archivos tienen.

3. Preprocesamiento de datos: Unificar formato de nombres de columnas, modificar los tipos de datos erroneos, tratar los valores ausentes y duplicados.

3. Análisis exploratorio de datos(EDA): Verificar correlación de variables númericas y categóricas, profundizar como influyen en la deserción de los clientes, verificar desbalance de clases, aplicacion de técnicas de sobremuestreo o submuestreo.

4. Entrenamiento y prueba del modelo: Preparar datos con eliminar las variables innecesarias, codificar variables categóricas(OHE), particionar los datos en conjuntos de entrenamiento(80%) y prueba(20%) y escalado de datos.

5. Aplicación y evaluación de modelos de ML: Crear función de evaluación de modelos con y sin ajuste de hiperparámetros mediante AUC-ROC, Accuracy y F1-Score, aplicar modelos como Logistic Regression, Decission Tree Classifier, Random Forest Classifier, XGBoost Classifier, LightGBM Classifier y CatBoost Classifier. 

6. Conlusión del proyecto: Elección del mejor modelo según sus métricas de desempeño.
