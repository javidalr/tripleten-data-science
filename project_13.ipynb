### Estacionariedad

Se realiza analisis de estacionariedad, para definir si los datos son estacionarios o no. Para iniciar, se revisa la desviacion estandar del conjunto de datos.

#### Desviacion Estandar

```python
# Verificacion de la desviacion estandar

data_day['rolling_std'] = data_day['num_orders'].rolling(10).std()
data_day.plot(figsize = (20,5))
plt.show()
```

La desviacion estandar de los datos son presuntamente constantes, lo que implica posibles datos estacionarios en t√©rminos de varianza, a pesar de una tendencia creciente de la media movil.

Para comprobar esto, se examinan las siguientes pruebas estad√≠sticas como la `Prueba Aumentada de Dickey-Fuller (ADF)` y la `Prueba de Kwiatkowski-Phillips-Schmidt-Shin (KPSS)` en el conjunto de datos original.

La hip√≥tesis nula (H0), en ambas pruebas, establece que la serie temporal tiene una ra√≠z unitaria, lo que significa que los datos no son estacionarios. La hip√≥tesis alternativa (H1), es que la serie temporal no tiene una ra√≠z unitaria y, por lo tanto, es estacionaria.

La presencia de una ra√≠z unitaria en datos de series temporales sugiere un comportamiento de camino aleatorio o no estacionariedad. Los datos no estacionarios suelen exhibir tendencias significativas o grandes fluctuaciones a lo largo del tiempo, lo que dificulta su predicci√≥n confiable mediante modelos tradicionales de series temporales.

#### Prueba Aumentada de Dickey-Fuller

```python
# Aplicacion de funcion `adf_test`

adf_test(data_1h)
```

#### Prueba de Kwiatkowski-Phillips-Schmidt-Shin

```python
# Aplicacion de funcion `kpss_test`
       
kpss_test(data_1h)
```

Seg√∫n los resultados de nuestras prueba, estos datos son estadisticamente estacionarios. Como resultado, se pueden aplicar m√©todos anal√≠ticos tradicionales como ARIMA (Auto-Regressive Integrated Moving Average) a estos datos.

### Modelos de Machine Learning

Para utilizar de forma correcta estos modelo, es necesario crear un nuevo conjunto de datos con caracteristicas adicionales, extraidas de las variables principales (`datetime` y `num_orders`)

#### Dataset con nuevas caracteristicas

```python
# Nuevas variables a partir de la columna 'datetime'

data_new = data_1h.copy()

data_new['year'] = data_new.index.year
data_new['month'] = data_new.index.month
data_new['day'] = data_new.index.day
data_new['dayofweek'] = data_new.index.dayofweek
data_new['quarter'] = data_new.index.quarter

data_new.head()
```

```python
# Nuevas variables a partir de la columna 'num_orders'

# Crear retardos hasta 6 horas
lags = [1, 2, 3, 4, 5, 6]
for lag in lags:
    data_new[f'lag_{lag}'] = data_new['num_orders'].shift(lag)

# Crear media movil con ventana de 3 a 5 horas
windows = [3, 4, 5]
for window in windows:
    data_new[f'rolling_mean_{window}'] = data_new['num_orders'].rolling(window = window, min_periods = 1).mean()

# Crear suma movil con ventana de 3 a 5 horas
for window in windows:
    data_new[f'rolling_sum_{window}'] = data_new['num_orders'].rolling(window = window, min_periods = 1).sum()

# Calculo media acumulativa
data_new['expanding_mean'] = data_new['num_orders'].expanding(1).mean()

# Eliminar filas de valores nulos
data_new = data_new.dropna()

display(data_new.head())
print('La estructura del nuevo conjunto de datos es ', data_new.shape)
```

#### Particion del nuevo dataset

```python
features = data_new.drop('num_orders', axis = 1)
target = data_new['num_orders']

features_train, features_test, target_train, target_test = train_test_split(features, target, shuffle = False, test_size = 0.1, random_state = seed)
```
