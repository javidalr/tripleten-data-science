### Tratamiento Valores Ausentes

```python
# Registros de los videojuegos sin antecedentes del nombre

games[games['name'].isnull()]
```

```python
# Registros sin antecedentes del a√±o de lanzamiento

games[games['year_of_release'].isnull()]
```

```python
# Eliminando registros del dataset, por falta de informaci√≥n

games.dropna(subset=['name','year_of_release'], axis=0, inplace=True)
```

```python
# Obtenemos los generos de video juegos en el dataset

games['genre'].unique()
```

```python
# Funci√≥n para obtener el promedio de calificaciones de critic_score y user_score

def calculate_average(df):
    
    # Agrupar por g√©nero y calcular el promedio de cada puntaje
    calc_avg = df.groupby('genre').agg({'critic_score': 'mean','user_score':'mean'}).reset_index()
    return calc_avg
```

```python
# calculando promedio de calificaci√≥n para cada g√©nero, y mostrando un dataframe con los valores

avg = calculate_average(games)
avg
```

```python
# Funci√≥n para reemplazar valores ausentes por la media de los g√©neros delos videojuegos 

def replace_values(df, promedios):
    
    # Diccionarios con los promedios de cada g√©nero
    avg_critic = promedios.set_index('genre')['critic_score'].to_dict()
    avg_user = promedios.set_index('genre')['user_score'].to_dict()
    
    # Reemplazar los valores ausentes en 'critic_score' y 'user_score'
    df['critic_score'] = df['critic_score'].fillna(df['genre'].map(avg_critic))
    df['user_score'] = df['user_score'].fillna(df['genre'].map(avg_user))
    return df
```

```python
# Reemplazando los valores de las medias en el dataframe

games = replace_values(games,avg)
games
```

```python
# Revisando el nuevo resumen del dataset

games.info()
```

**Tratamiento valores ausentes**

En primer lugar se revisaron los valores ausentes en la columna 'name', que a su vez esos registros tenian mas valores nulos, adicionando los de la columna 'genre', adem√°s se observaron los nulos de la columna 'year_of_release', y por ser una informaci√≥n faltante muy relevante se defini√≥ eliminar todos estos registros del dataset.

Luego, en la columna 'user_score', por ser una gran cantidad de datos sin informaci√≥n, se calcularon las calificaciones promedio (media) para cada g√©nero, y se reemplazaron estos valores nulos por la calificaci√≥n promedio al genero que corresponde el t√≠tulo.

Por √∫ltimo, los valores nulos de la columna 'rating' no se manipularan por el momento, ya que no es posible obtener o predecir la clasificaci√≥n ESRB para los t√≠tulos que no la poseen.

### Ingenier√≠a de Caracter√≠sticas (Feature Engineering)

```python
# Agregando columna 'total_sales' con las ventas totales de cada juego

games['total_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
```

```python
# Funci√≥n que genera categorias de la columna 'critic_score'

def critic_category(grade):
    if grade < 20:
        return 'very bad'
    elif grade < 40:
        return 'bad'
    elif grade < 60:
        return 'regular'
    elif grade < 80:
        return 'good'
    else:
        return 'very good'
```

```python
# Agregando columna 'critic_category', usando fx critic_category con metodo apply

games['critic_category'] = games['critic_score'].apply(critic_category)
```

```python
# Funci√≥n que genera categorias de la columna 'user_score'

def user_category(grade):
    if grade < 2:
        return 'very bad'
    elif grade < 4:
        return 'bad'
    elif grade < 6:
        return 'regular'
    elif grade < 8:
        return 'good'
    else:
        return 'very good'
```

```python
# Agregando columna 'user_category', usando fx user_category con metodo apply

games['user_category'] = games['user_score'].apply(user_category)
```

## **Pruebas de hip√≥tesis**

### Hip√≥tesis 1: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

```python
# Prueba de hip√≥tesis

print('Formulaci√≥n de Prueba de Hipotesis')
print()
print('H0: Las calificaciones promedio de los usuarios de las plataformas Xbox One y PC son iguales.')
print('H1: Las calificaciones promedio de los usuarios de las plataformas Xbox One y PC son diferentes.')

xone = games_filter_2006[games_filter_2006['platform']=='XOne']['user_score']
pc = games_filter_2006[games_filter_2006['platform']=='PC']['user_score']
print()

alpha = 0.01
print('Nivel de significancia = 1%')
print()

resultado = st.ttest_ind(xone, pc)
print(f'El valor de p_value es: {resultado.pvalue}')
print()
print()
print('Conclusi√≥n')
print()

if resultado.pvalue < alpha:
    print('Como el pvalue es menor al nivel de significancia de la prueba, se rechaza la hip√≥tesis nula, ya que hay evidencia\nestad√≠stica, con un nivel de confianza del 99%, que las calificaciones promedio de los usuarios de las plataformas\nXbox One y PC son diferentes.')
else:
    print('Como el pvalue es mayor al nivel de significancia de la prueba, se aprueba la hip√≥tesis nula, ya que hay evidencia\nestad√≠stica, con un nivel de confianza del 99%, que las calificaciones promedio de los usuarios de las plataformas\nXbox One y PC son iguales.')
```

### Hip√≥tesis 2: Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes.

```python
# Prueba de hip√≥tesis

print('Formulaci√≥n de Prueba de Hipotesis')
print()
print('H0: Las calificaciones promedio de los usuarios sobre los generos de Acci√≥n y Deportes en videojuegos son iguales.')
print('H1: Las calificaciones promedio de los usuarios sobre los generos de Acci√≥n y Deportes en videojuegos son diferentes.')

action = games_filter_2006[games_filter_2006['genre']=='Action']['user_score']
sport = games_filter_2006[games_filter_2006['genre']=='Sports']['user_score']
print()
alpha = 0.01
print('Nivel de significancia = 1%')
print()

resultado = st.ttest_ind(action, sport)
print(f'El valor de p_value es: {resultado.pvalue}')
print()
print()
print('Conclusi√≥n')
print()

if resultado.pvalue < alpha:
    print('Como el pvalue es menor al nivel de significancia de la prueba, se rechaza la hipotesis nula, ya que hay evidencia\nestad√≠stica, con un nivel de confianza del 99%, que las calificaciones promedio de los usuarios sobre los generos\nde Acci√≥n y Deportes en videojuegos son diferentes.')
else:
    print('Como el pvalue es mayor al nivel de significancia de la prueba, se aprueba la hipotesis nula, ya que hay evidencia\nestad√≠stica, con un nivel de confianza del 99%, que las calificaciones promedio de los usuarios sobre los generos\nde Acci√≥n y Deportes en videojuegos son iguales.')
```
